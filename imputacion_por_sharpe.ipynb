Imputación de nans por sharpe a una serie

```python
import numpy as np

def impute_sharpe_closed(series, T=None):
    """
    Imputa en la serie (con np.nan en los datos faltantes) un único valor z,
    de forma que el Sharpe ratio global (media/σ, con σ la desviación
    estándar poblacional) se iguale al Sharpe ratio obtenido de los datos conocidos.

    Parámetros:
      - series: lista o arreglo de números (np.nan donde falte el dato).
      - T: (opcional) Sharpe ratio objetivo. Si es None, se calcula con los datos conocidos.

    Retorna:
      - imputed_series: arreglo con los valores faltantes reemplazados por z.
      - z: valor imputado.
      - T: Sharpe ratio usado.
    """
    # Convertir a array float y obtener máscara de faltantes
    series = np.array(series, dtype=float)
    missing_mask = np.isnan(series)
    known = series[~missing_mask]

    k = len(known)         # número de datos conocidos
    m = np.sum(missing_mask)  # número de faltantes
    n = k + m              # tamaño total de la serie

    if k == 0:
        raise ValueError("La serie no tiene datos conocidos para calcular el Sharpe ratio.")

    # Suma y suma de cuadrados de los datos conocidos
    A = np.sum(known)
    S2 = np.sum(known**2)

    # Calcular el Sharpe ratio objetivo con los datos conocidos, si no se proporciona
    if T is None:
        mu_known = A / k
        sigma_known = np.std(known, ddof=0)  # desviación estándar poblacional
        if sigma_known == 0:
            raise ValueError("La desviación estándar de los datos conocidos es cero.")
        T = mu_known / sigma_known

    # Coeficientes de la ecuación cuadrática
    a = m * (m - T**2 * (n - m))
    b = 2 * A * m * (1 + T**2)
    c = A**2 * (1 + T**2) - T**2 * n * S2

    # Resolver la ecuación cuadrática
    coeffs = [a, b, c]
    sols = np.roots(coeffs)  # pueden haber dos soluciones (reales o complejas)

    # Seleccionar las soluciones reales
    sols = sols[np.isreal(sols)].real
    if len(sols) == 0:
        raise ValueError("No se encontró solución real para la imputación.")

    # Elegir la solución más cercana a la media de los datos conocidos
    z = sols[np.argmin(np.abs(sols - np.mean(known)))]

    # Imputar el valor z en los lugares faltantes de la serie
    imputed_series = series.copy()
    imputed_series[missing_mask] = z

    return imputed_series, z, T

def iterative_imputation(series):
    """
    Recorre la serie desde el inicio hasta el final e, de forma iterativa,
    cuando se encuentra un valor faltante (np.nan), se calcula su imputación
    usando la función impute_sharpe_closed en el segmento que va desde el
    primer elemento hasta el valor faltante actual. Luego, se reemplaza dicho
    valor en la serie y se continúa con el resto.

    Parámetros:
      - series: lista o arreglo de números (con np.nan donde falte el dato).

    Retorna:
      - series_imputada: arreglo con todos los valores faltantes imputados.
    """
    series = np.array(series, dtype=float)
    nans_contador = 0
    # Recorrer la serie
    for i in range(len(series)):
        if np.isnan(series[i]):
            # Tomar el segmento desde el inicio hasta el índice i (inclusive)
            seg = series[:i+1]
            # Imputar en este segmento (en el que normalmente habrá un solo nan)
            seg_imputed, z, T = impute_sharpe_closed(seg)
            # Actualizar la posición i con el valor imputado
            series[i] = z
            nans_contador += 1
            # Se puede imprimir información para seguimiento (opcional)
            # print(f"Imputando posición {i} con valor {z:.4f} (Sharpe objetivo: {T:.4f})")
    print(nans_contador)
    return series

serie = x # Tiene q ser un tipo de dato np.array
serie_imputada = iterative_imputation(serie)
# print("Serie imputada iterativamente:", serie_imputada)
```

Código para calcular el sharpe

```python
import numpy as np

def calculate_sharpe(series, rf=0.0):
    """
    Calcula el Sharpe ratio de una serie, ignorando los valores faltantes (np.nan).

    Parámetros:
      - series: lista o arreglo de números (puede contener np.nan).
      - rf: tasa libre de riesgo (default=0.0).

    Retorna:
      - sharpe: el Sharpe ratio, calculado como (media - rf) / σ.
    """
    # Convertir a array y eliminar los np.nan
    series = np.array(series, dtype=float)
    cleaned_series = series[~np.isnan(series)]

    # Calcular media y desviación estándar (poblacional, ddof=0)
    mu = np.mean(cleaned_series)
    sigma = np.std(cleaned_series)

    if sigma == 0:
        raise ValueError("La desviación estándar es cero, no se puede calcular el Sharpe ratio.")

    sharpe = (mu - rf) / sigma
    return sharpe

```
